# read

```python
import pandas as pd; pd.set_option('display.max_columns', None)
import numpy as np
import warnings; warnings.filterwarnings(action='ignore')

train = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/dacon_job/data/train.csv')
test  = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/dacon_job/data/test.csv')
sub   = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/dacon_job/data/sample_submission.csv')

att_d = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/dacon_job/data/속성_D_코드.csv').iloc[:, :-1]
att_h = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/dacon_job/data/속성_H_코드.csv')
att_l = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/dacon_job/data/속성_L_코드.csv')
```

```python
for i in [1, 2, 3]:
    att_d_copy = att_d.copy()
    att_d_copy.columns = [f'prefer_{j}_{i}' for j in att_d.columns]
    train = train.merge(att_d_copy, how='left', left_on=f'person_prefer_d_{i}', right_on=f'prefer_속성 D 코드_{i}')

for i in [1, 2, 3]:
    att_h_copy = att_h.copy()
    att_h_copy.columns = [f'prefer_{j}_{i}' for j in att_h.columns]
    train = train.merge(att_h.copy, how='left', left_on=f'person_prefer_h_{i}', right_on=f'prefer_속성 H 코드_{i}')
```

```python
att_d_copy = att_d.copy()
att_d_copy.columns = [f'content_{j}' for j in att_d.columns]
train = train.merge(att_d_copy, how='left', left_on='contents_attribute_d', right_on='content_속성 D 코드')
```

# drop

```python
drop_cols = ['id', 'contents_open_dt']
train.drop(drop_cols, axis=1, inplace=True)
test.drop(drop_cols, axis=1, inplace=True)
```

# modeling

```python
X = train.drop('target', axis=1)
y = train['target']
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(
    n_estimators=300,
    max_depth=60, 
    n_jobs=-1
)

model.fit(X, y)
```

# sub

```python
sub['target'] = model.predict(test)
sub.head()
```
