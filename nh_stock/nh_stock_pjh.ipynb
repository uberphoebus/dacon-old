```python
for i in range(train_count.shape[0]):
    
    count_idx = train_count.index[i]
    
    # 계좌/종목을 날짜기준으로 정렬
    train_sort = train[
        (train['act_id'] == count_idx[0]) & (train['iem_cd'] == count_idx[1])
    ].sort_values('byn_dt')
    
    # 해당 계좌와 종목 인덱스
    act_id = count_idx[0]
    iem_cd = count_idx[1]
    
    # 첫 번째 거래는 해당 hold_d를 past_d로 동일하게 부여
    train['past_d'][
        (train['act_id'] == act_id) & 
        (train['iem_cd'] == iem_cd) & 
        (train['byn_dt'] == train_sort.iloc[0, 2])
    ] = train_sort.iloc[0, 3]
    
    # 이후의 거래는 한 단계 이전 거래의 hold_d를 past_d로 부여
    for j in range(train_sort.shape[0] - 1):
        
        sort_idx = j
        byn_dt = train_sort.iloc[sort_idx + 1, 2]
        
        train['past_d'][
            (train['act_id'] == act_id) & (train['iem_cd'] == iem_cd) & (train['byn_dt'] == byn_dt)
        ] = train_sort.iloc[sort_idx, 3]
    
    if i % 5000 == 0:
        print(i, 'done')
train.head(30)
```
