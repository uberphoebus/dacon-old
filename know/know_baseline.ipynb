# read

```python
import pandas as pd
from glob import glob
import warnings; warnings.filterwarnings(action='ignore')

train = [pd.read_csv(path) for path in sorted(glob('./data/train/*.csv'))]
test  = [pd.read_csv(path) for path in sorted(glob('./data/test/*.csv'))]
```

```python
for df1, df2 in zip(train, test):
    print(f'train {df1.shape} / test {df2.shape}')
```

# 전처리

```python
for df in train:
    for col in df.columns:
        df[col].replace(' ', '0', inplace=True)

for df in test:
    for col in df.columns:
        df[col].replace(' ', '0', inplace=True)
```
