## lightgbm

```python
import lightgbm as lgb
import pandas as pd
```

```python
## data갖고오기
data=pd.read_csv('./data/train.csv', encoding='cp949')
```

## Data Preprocessing
#### 데이터 원본 데이터로 복구해서 사용

```python
data['round_send']=data['SEND_SPG_INNB'].round(-10)
data['round_rec']=data['REC_SPG_INNB'].round(-10)
```

```python
data
```

#### 격자공간을 나타내는 수는 categorical하다고 생각하여 {격자 숫자 : num} 로 dictionary 생성
#### 수신, 송신 격자 모두 같은 dictionary에 넣어 중복 제거했습니다.
#### (1129 과 5011은 다른 격자 공간을 의미할 뿐, 크고 작은 수치가 아니기 때문)

```python
a=list(set(data.round_send.to_list()))
```

```python
b=list(set(data.round_rec.to_list()))
```

```python
sums=list(set(a+b))
```

```python
len(sums)
```

```python
nn=[]
for _ in range(len(sums)):
    n=[sums[_],_]
    nn.append(n)
```

```python
nn=dict(nn)
```

### one hot encoding

```python
from tqdm import tqdm
for _ in tqdm(range(32000)):
    data.loc[_,'send_dict']=str(nn[data['round_send'][_]])
    data.loc[_,'rec_dict']=str(nn[data['round_rec'][_]])
```

```python
data=pd.get_dummies(data)
```

```python
data
```

### 총 feature 511개

```python
feature=list(data.columns)[6:]
```

```python
len(feature)
```

```python
data[feature].head()
```

### train-test split, random state 42고정

```python
from sklearn.model_selection import train_test_split
raw_x=data[feature]
raw_y=data[['INVC_CONT']]

train_x, test_x, train_y, test_y = train_test_split(raw_x, raw_y, test_size=0.2, random_state=42)

train_ds = lgb.Dataset(train_x, label = train_y) 
test_ds = lgb.Dataset(test_x, label = test_y) 

print(train_y.max())
print(test_y.max())
```

## model training

```python

params = { 
          'learning_rate': 0.082, 
          'max_depth': 5, 
          'boosting': 'gbdt', 
          'objective': 'regression', 
          'metric': 'mse', 
          'is_training_metric': True, 
          'num_leaves': 50, 
          'feature_fraction': 0.8, 
          'bagging_fraction': 0.6, 
          'bagging_freq': 2, 
         
          'min_data_in_leaf':2
    ,
         
         }

model = lgb.train(params, train_ds, 1500, test_ds, verbose_eval=100, early_stopping_rounds=200 )
```

```python
from sklearn.metrics import mean_squared_error

mean_squared_error(test_y,model.predict(test_x))
```
