```python
import pandas as pd
train = pd.read_csv('./data/train.csv')
test  = pd.read_csv('./data/test.csv')
```

```python
train.shape, test.shape
```

```python
train.dtypes
```

```python
train_cols = ['id', 'Year Built', 'Year Remod/Add', 'Garage Yr Blt',  'Overall Qual', 'Exter Qual', 'Bsmt Qual', 'Kitchen Qual', 'Gr Liv Area', 'Total Bsmt SF', '1st Flr SF', 'Garage Area', 'Full Bath', 'Garage Cars', 'target']
test_cols = ['id', 'Year Built', 'Year Remod/Add', 'Garage Yr Blt',  'Overall Qual', 'Exter Qual', 'Bsmt Qual', 'Kitchen Qual', 'Gr Liv Area', 'Total Bsmt SF', '1st Flr SF', 'Garage Area', 'Full Bath', 'Garage Cars']

train = train[train_cols]
test = test[test_cols]
train.head()
```

```python
year_cols = train_cols[1:4]

train[year_cols] = 2022 - train[year_cols]
test[year_cols] = 2022 - test[year_cols]

train.head()
```

```python
for df in [train, test]:
    for col in [['Gr Liv Area', 'Total Bsmt SF', '1st Flr SF', 'Garage Area']]:
        q1 = df[col].quantile(.25)
        q3 = df[col].quantile(.75)
        iqr = q3 - q1
        low = q1 - 1.5 * iqr
        high = q3 + 1.5 * iqr
        df.loc[df[col] < low, col] = low
        df.loc[df[col] < high, col] = high
```
